```python
import alpaca_trade_api as tradeapi
from config import ALPACA_CONFIG

alpaca = tradeapi.REST(**ALPACA_CONFIG)

def calcular_cantidad_riesgo(precio_activo, porcentaje_riesgo=0.01):
    try:
        cuenta = alpaca.get_account()
        equity_total = float(cuenta.equity)
        dinero_a_invertir = equity_total * porcentaje_riesgo
        return round(dinero_a_invertir / precio_activo, 6)
    except: return 0.0

def ejecutar_compra_alpaca(ticker, cantidad_calculada):
    try:
        symbol = ticker.upper() + "/USD"
        if cantidad_calculada > 0:
            alpaca.submit_order(symbol=symbol, qty=cantidad_calculada, side='buy', type='market', time_in_force='gtc')
            return True
    except Exception as e:
        print(f"‚ùå Error Compra {ticker}: {e}")
    return False

def liquidar_todo_el_ticker(ticker, motivo):
    try:
        symbol = ticker.upper() + "/USD"
        alpaca.close_position(symbol)
        print(f"üö® [VENTA] {ticker.upper()} liquidado por {motivo}")
        return True
    except:
        try: # Reintento sin diagonal
            alpaca.close_position(ticker.upper() + "USD")
            return True
        except: return False
```
