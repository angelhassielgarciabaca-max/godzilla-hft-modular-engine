```python
import time
from datetime import datetime
from collections import deque
import motores_data
import cvd_g

# --- CONFIGURACI√ìN INICIAL ---
WATCHLIST = ['btc', 'eth', 'sol', 'bnb']
CAPITAL_INICIAL = 100000 # Tu balance de $100k
# Buffers de datos (Deltas y Volumen Total)
flujo_deltas = {t: deque(maxlen=1000) for t in WATCHLIST}
volumen_total = {t: deque(maxlen=1000) for t in WATCHLIST}

# Estado del Mercado y Posiciones
ESTADO_MERCADO = {t: {
    "en_referencia": False, 
    "fase_ciclo": None, 
    "posicion_activa": False,
    "precio_compra": 0.0  # Nuevo: Para calcular utilidad
} for t in WATCHLIST}

def imprimir_leyenda():
    print("\n" + "="*50)
    print("üî• GODZILLA v5.0 ENGINE DESPERT√ì Y EST√Å ATACANDO...")
    print(f"‚è∞ Inicio de cacer√≠a: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
    print("üéØ Objetivo: $100k Account Protection & Delta Efficiency")
    print("="*50 + "\n")

def ejecutar_orden(ticker, tipo, precio):
    ahora = datetime.now().strftime('%H:%M:%S.%f')[:-3]
    estado = ESTADO_MERCADO[ticker]
    
    if tipo == "buy":
        estado["posicion_activa"] = True
        estado["precio_compra"] = precio
        print(f"[{ahora}] üü¢ COMPRA EJECUTADA: {ticker.upper()} a ${precio:,.2f}")
    
    elif tipo == "sell":
        # C√°lculo de utilidad
        ganancia_usd = precio - estado["precio_compra"]
        porcentaje = (ganancia_usd / estado["precio_compra"]) * 100
        color = "‚úÖ" if ganancia_usd > 0 else "üö®"
        
        print(f"[{ahora}] üî¥ VENTA EJECUTADA: {ticker.upper()} a ${precio:,.2f}")
        print(f"    {color} RESULTADO: ${ganancia_usd:,.2f} ({porcentaje:.2f}%)")
        
        estado["posicion_activa"] = False
        estado["precio_compra"] = 0.0

# --- LOOP PRINCIPAL ---
imprimir_leyenda()

# Aqu√≠ iniciar√≠as el thread de motores_data.iniciar_socket(WATCHLIST)

while True:
    for ticker in WATCHLIST:
        # 1. Obtener datos actuales del mercado (Simulado o desde API)
        precio_actual = obtener_precio_actual(ticker) # Tu funci√≥n de precio
        
        # 2. Validar fuerza con el nuevo m√≥dulo de Eficiencia
        # Pasamos AMBOS buffers: deltas y volumen
        decision = cvd_g.validar_fuerza_ciclo(
            ticker, 
            flujo_deltas[ticker], 
            volumen_total[ticker]
        )
        
        # 3. Ejecuci√≥n basada en diagn√≥stico
        if decision == "EJECUTAR_COMPRA":
            ejecutar_orden(ticker, "buy", precio_actual)
            
        elif decision == "EJECUTAR_VENTA":
            ejecutar_orden(ticker, "sell", precio_actual)
            
    time.sleep(0.5) # Ciclo de alta frecuencia
```
